**Q1: What is linear regression?** 

It is a supervised machine learning algorithm that finds the linear relationship between independent and dependent variables by minimizing the sum of squared residuals by using ordinary least square method(OLS) or Gradient descent.



**Q2: How to iterpret the slope and coefficients in linear regression?**

**Slope Coefficient(β1):**- It represents the estimated increase in y for every one unit increase in x1 keeping other variables constant.

**Intercept (β0):**- It represents the estimated value of y when all the independent variables are set to zero.



**Q3: What are the assumptions of a linear model?**

There are 3 types of assumptions:
1. Assumptions about form of the data
    *  **Linear relationship between X and y.**
    
            How to check assuption:
                * Create a scatter plot for feature and target.
                * If you have multiple features create a scatterplot for   predicted and actual values.
                
            What it will affect:
                * Predictions will be extremly inaccurate because of  underfitting. Thiis is a seirous violation that should                   not be ignored.
            
            How to fix:
                * Apply a non-linear transformation (ex : log-transformation)
                * Add polynomial terms(ex: add X**2 term if the X ,y plot has parabolic shape.)
      
2. Assumptions about errors/residuals:
    * **Normality :** Residuals normally distributed with the mean of 0.
        
            How to check assuption:
                * Histogram
                * Q-Q Plot
                * Shaphiro wilk normality test.
                * Check omnibus (test that describe normality of residuals.)
            What it will affect:
            How to fix:
                * Handle with  outliers.
                * Log- transformation to the target variable.
      
    * **Homoscedasticity :** It is assumed that the residual errors have the same variance. 
        
            How to check assuption: The easiest way to check for this assumption is to create a residual versus fitted value plot.
            What it will affect:
            How to fix:
                * Log transformation.
                * Redefine target variable --> use rates instead of raw value (ex: instead of measurinf # of school based on population, predict school per person.)
                * Use weighted regression which assigns weight to each data point based on its predicted value.(Gives small weight data with higher variance.)
      
    * **Independence :**All independent variables are uncorrelated with the error term (residuals).
        
            How to check assuption:
                * Perform Durbin-Watson Test.
                    dw = 2 --> no auto correlation
                    dw>2 --> positive serial auto correlation 
                    dw<2 --> negative serial auto correlation
                * Create a plot of residuals and predicted values.(?? - It should follow inioform distributions.)
            What it will affect:
                * It will affect model estimations. Estimated coefficients might have incorrect signs.
            How to fix:
                * For positive autocorrelation add logs of the predictor variables.
                * For seasonal correlation, consider adding seosonal dummy variables.
                
 
3. Assumptions about estimators: 
    * **No multicolinearity between estimastors.**
    
            How to check assuption:
                * Check correlation matrix or heatmap. 
                * Calculate VIF(VarianceInflation Factor)
                    If VIF > 5 --> potential multicolinearity.
            What it will affect:
                * Cause an issues with interpretion of coeffieients.
                * Increase standard error of the coefs. 
            How to fix:
                * Remove predictor with high VIF score.
                * Apply dimensinality reduction like PCA.
                * Alternatively we can use LASSO or RIDGE regression. They are designed to handle predictors variables that are highly correlated.
                

**Q4 : What are the evaluation metrics in linear regression?**

* MAE --> Robust to outliers, 
    * easy to interpret because error in the same unit.
    * Time consuming.
    * Can't take derivetive at 0 , cant converge with gradient decsent.
* MSE --> 
    * Equation is differantiable and convew(has global min)
    * Not trobust to outliers.
    * Penalize error by squaring it.
    * Error in the squared unit and not easy to interpret.
* RMSE --> 
    * Error in the same unit.
    * Penalize error less than MSE, and more than MAE
* R2 -->
    * It always increases when you keep adding new predictors.
    * It used for explaining how well the independent variables in the linear regression model explains the variability in the dependent variable.
   
 
* Adj R2 --> 
    * It will not increase when you add useeless predictors.
    * It used for explaining how well the independent variables in the linear regression model explains the variability in the dependent variable. 
    




**NOTES:**
- Use RMSE when you want to penalize error.
- Use MAE if you want more interpretable results.
- Use Adj R2 when you have multiple predictors. Because when you add more and more predictors R2  always will increase even if you add a useless predictor. It may lead you think that model is doing well but it will not work with the test data well.)
- For R2 and Adj R2 higher scores are better. But for rest lower are better.

**Q7: What is a Gradient Descent ?**

Gradient Descent is an iterative optimization algorithm use to find a local minimum/maximum of a given function.
The goal of gradient descent is to minimize the error/loss between predicted and actual y. It requires direction and learning rate(referred as step size or alpha)to arrive at the local or global minimum. A high learning rate results in larger steps but there is a risk of overshooting the minimum while low learning rate or small steps have the advantage of more precision as this takes more time and computations to reach the minimum.

**Q8: What are the challenges in Gradient Descent?**
Gradient Descent can find global minimum with ease however non-convex problems struggles to find a global minimum. Noisy gradients can help escape local minimums.

**Q9:  How to interpret a Q-Q plot in linear regression?**

A normal probability plot, or more specifically a quantile-quantile (Q-Q) plot, shows the distribution of the data against the expected normal distribution. Q -Q plot helps find the normality of the residual/error term.
